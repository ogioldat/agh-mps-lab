#### Today's problem

What is the density of the room air?

#### Ideal gas
- $pV=nRT$

#### The model we're gonna use
- $\rho =  pR'T$
  - $\rho$, density
  - $p$, pressure
  - $R'$, gas constant, $\frac{R}{M}$,
    - $R$, molar gas constant,
    - $M$, air mass molar
  - $T$, temperature

```python
import pint
from scipy import constants
import matplotlib.pyplot as plt
import numpy as np
```

```python
si = pint.UnitRegistry()

p = 1000 * si.hPa
p += 3 * si.N / si.cm**2

p
```

```python
p = 1000 * si.hPa

R = constants.R * si.J / si.mol / si.K
M = 0.0289 * si.kg / si.mol
Rp = R/M
Rp
```

```python
T = (17 + constants.zero_Celsius) * si.K
rho = p / Rp / T
rho

rho.check("[speed]")
```

```python
rho.check("[density]")
```

```python
rho.to_base_units()
```

```python
f'{rho =: .3g}'
```

```python
si.setup_matplotlib()
```

```python
def rho(T):
    return p / Rp / T

temp = np.linspace(250, 320) * si.degK

plt.plot(temp, rho(temp))
```

```python
plt.gca().xaxis.set_units(si.degC)
plt.gca().yaxis.set_units(si.kg / si.m**3)
plt.plot(temp, rho(temp))
```
